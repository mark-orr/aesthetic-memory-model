# Exploring f(x, r) = 1 - (x/r)^2

A small playground for `inverted_parabola` (`src/utils.py`): an inverted parabola peaking at `f(0, r) = 1`, crossing zero at `x = ±r`, and going unboundedly negative beyond that. `r` sets the width of the positive hump. Preliminary exploration -- not yet tied to a specific Design or algorithm.

Change `R` and `X_RANGE` below and re-run to explore fixed-r/varying-x behavior; the second plot overlays several values of `r` so you can see how the hump width changes.

```python
import sys
sys.path.insert(0, "..")
import numpy as np
import matplotlib.pyplot as plt

from src.utils import inverted_parabola

R = 3.0            # fixed r for the main plot
X_RANGE = 2.0       # x spans [-X_RANGE*R, +X_RANGE*R]
```

## Fixed r, varying x

```python
x = np.linspace(-X_RANGE * R, X_RANGE * R, 400)
y = inverted_parabola(x, R)

fig, ax = plt.subplots(figsize=(7, 4.5))
ax.plot(x, y)
ax.axhline(0, color="gray", linewidth=0.5)
ax.axvline(0, color="gray", linewidth=0.5)
ax.axvline(R, color="red", linestyle="--", linewidth=0.8, label=f"x=\u00b1r (r={R})")
ax.axvline(-R, color="red", linestyle="--", linewidth=0.8)
ax.set_xlabel("x")
ax.set_ylabel("f(x, r) = 1 - (x/r)^2")
ax.set_title(f"f(x, r) = 1 - (x/r)^2, fixed r={R}")
ax.legend()
fig.tight_layout()
fig.savefig("../results/figures/inverted_parabola_fixed_r.png", dpi=150)
```

## How r changes the shape

```python
R_VALUES = [1.0, 2.0, 3.0, 5.0]
x = np.linspace(-6, 6, 400)

fig, ax = plt.subplots(figsize=(7, 4.5))
for r in R_VALUES:
    ax.plot(x, inverted_parabola(x, r), label=f"r={r}")
ax.axhline(0, color="gray", linewidth=0.5)
ax.axvline(0, color="gray", linewidth=0.5)
ax.set_xlabel("x")
ax.set_ylabel("f(x, r)")
ax.set_title("f(x, r) = 1 - (x/r)^2 for several r")
ax.legend()
fig.tight_layout()
fig.savefig("../results/figures/inverted_parabola_varying_r.png", dpi=150)
```
