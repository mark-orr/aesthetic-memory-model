# Aesthetic basis vs. complexity

Reads `results/data/aesthetic_basis.csv` (generated by `run_simulation.py`) and plots the Algorithm A `aesthetic_basis` (`|predicted activation - actual activation|`) against each song's `complexity`.

**Caveat:** the current toy `SONG_SEQUENCE` in `run_simulation.py` only produces a handful of repeat-exposure trials across 3 distinct complexity values, so this is a pipeline smoke test, not yet a real test of the inverted-U hypothesis. Re-run `run_simulation.py` with a richer, more repeated song sequence (or real stimulus data) before drawing scientific conclusions from the shape of this plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../results/data/aesthetic_basis.csv")
df
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(df["complexity"], df["aesthetic_basis"])
for _, row in df.iterrows():
    ax.annotate(row["actual_song_id"], (row["complexity"], row["aesthetic_basis"]),
                textcoords="offset points", xytext=(5, 5))
ax.set_xlabel("complexity")
ax.set_ylabel("aesthetic_basis")
ax.set_title("Aesthetic basis vs. complexity")
fig.tight_layout()
fig.savefig("../results/figures/aesthetic_basis_vs_complexity.png", dpi=150)
```
