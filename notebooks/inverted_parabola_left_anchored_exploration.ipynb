# Exploring the left-anchored inverted parabola: f(x, r) = 1 - (2x/r - 1)^2

A playground for `inverted_parabola_left_anchored` (`src/utils.py`): an inverted parabola whose left x-intercept is fixed at `x=0`, whose vertex height is fixed at `1` (at `x=r/2`), and whose right x-intercept sits at `x=r` -- so `r` directly parameterizes the full width of the hump on the x-axis. Preliminary exploration, follow-on from `inverted_parabola_exploration.ipynb`.

Change `R` and re-run to explore fixed-r/varying-x behavior; the second plot overlays several values of `r` so you can see how the width (and the location of the vertex/right root) changes.

```python
import sys
sys.path.insert(0, "..")
import numpy as np
import matplotlib.pyplot as plt

from src.utils import inverted_parabola_left_anchored

R = 4.0   # full width on the x-axis: left root at 0, right root at R
```

## Fixed r, varying x

```python
x = np.linspace(-0.5 * R, 1.5 * R, 400)
y = inverted_parabola_left_anchored(x, R)

fig, ax = plt.subplots(figsize=(7, 4.5))
ax.plot(x, y)
ax.axhline(0, color="gray", linewidth=0.5)
ax.axvline(0, color="red", linestyle="--", linewidth=0.8, label="left root (x=0)")
ax.axvline(R, color="red", linestyle="--", linewidth=0.8, label=f"right root (x=r={R})")
ax.axvline(R / 2, color="green", linestyle=":", linewidth=0.8, label="vertex (x=r/2)")
ax.set_xlabel("x")
ax.set_ylabel("f(x, r) = 1 - (2x/r - 1)^2")
ax.set_title(f"Left-anchored inverted parabola, fixed r={R}")
ax.legend()
fig.tight_layout()
fig.savefig("../results/figures/inverted_parabola_left_anchored_fixed_r.png", dpi=150)
```

## How r changes the shape

```python
R_VALUES = [1.0, 2.0, 4.0, 8.0]
x = np.linspace(-1, 9, 400)

fig, ax = plt.subplots(figsize=(7, 4.5))
for r in R_VALUES:
    ax.plot(x, inverted_parabola_left_anchored(x, r), label=f"r={r}")
ax.axhline(0, color="gray", linewidth=0.5)
ax.axvline(0, color="gray", linewidth=0.5)
ax.set_ylim(-3, 1.2)
ax.set_xlabel("x")
ax.set_ylabel("f(x, r)")
ax.set_title("Left-anchored inverted parabola for several r (all start at x=0)")
ax.legend()
fig.tight_layout()
fig.savefig("../results/figures/inverted_parabola_left_anchored_varying_r.png", dpi=150)
```
